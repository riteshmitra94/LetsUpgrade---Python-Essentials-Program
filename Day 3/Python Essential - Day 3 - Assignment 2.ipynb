```python
marks = int(input('Input your marks out of 100 '))

if marks >=80:
    print('That\'s a Distinction!')
elif marks < 80 and marks >= 75:
    print('That\'s a First class!')
elif marks < 75 and marks >=60:
    print('That\'s a Second class')
elif marks < 60 and marks > 40:
    print('That\'s a Third class')
else:
    print('You need to work harder')
```

```python
def isPrime(n):
    
    if (n <= 1):
        return False
    for i in range(2, n):
        if (n % i == 0):
            return False
 
    return True

primes = []

for i in range(1, 1000):
    if isPrime(i):
        primes.append(i)

print('Prime Numbers from 1 to 1000 are:\n ',primes)
```
